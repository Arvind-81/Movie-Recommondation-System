# Cleaning Dataset

```python
#Dropping Column
data_df=data_df.drop(['title_x','imdb_id','poster_path','wiki_link','title_y','is_adult','imdb_votes','story','summary','tagline','wins_nominations','release_date'],axis=1)
data_df
```

```python
#Replacing nonnumeric value to Nan
data_df['runtime']=pd.to_numeric(data_df['runtime'],errors='coerce')
```

```python
#Dropping NaN value with '-'
data_df['runtime']=data_df['runtime'].fillna('')
```

```python
data_df.tail()
```

```python
data_df['year_of_release']=pd.to_numeric(data_df['year_of_release'],errors='coerce')
```

```python
data_df['year_of_release']=data_df['year_of_release'].fillna('')
```

```python
data_df['imdb_rating']=pd.to_numeric(data_df['imdb_rating'],errors='coerce')
```

```python
data_df['imdb_rating']=data_df['imdb_rating'].fillna('')
```

```python
data_df['actors']=data_df['actors'].fillna('-')
```
